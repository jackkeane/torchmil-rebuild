# CAMELYON16 Tutorial (torchmil-rebuild)

End-to-end example: load dataset, train ABMIL, and evaluate.

```python
from torch.utils.data import DataLoader
import torch
from torchmil.datasets import Camelyon16MIL
from torchmil.data import mil_collate_fn
from torchmil.models import ABMIL
from torchmil.utils import Trainer
```

```python
train_ds = Camelyon16MIL(root='../data', features='UNI', split='train', download=True)
test_ds = Camelyon16MIL(root='../data', features='UNI', split='test', download=False)
train_loader = DataLoader(train_ds, batch_size=1, shuffle=True, collate_fn=mil_collate_fn)
test_loader = DataLoader(test_ds, batch_size=1, shuffle=False, collate_fn=mil_collate_fn)
print(train_ds.data_dim, len(train_ds), len(test_ds))
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = ABMIL(in_shape=train_ds.data_dim, num_classes=2)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
trainer = Trainer(model=model, optimizer=optimizer, device=device)
logs = trainer.train(train_loader, epochs=5, val_dataloader=test_loader)
logs[-1]
```
